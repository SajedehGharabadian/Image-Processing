```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('input/SuperMan.jpg')
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
```

```python
img_hsv = cv2.cvtColor(img_rgb,cv2.COLOR_RGB2HSV)

H, S, V = cv2.split(img_hsv)
```

```python
background = cv2.imread('input/sky.jpg')
background = cv2.cvtColor(background,cv2.COLOR_BGR2RGB)
plt.imshow(background)
```

```python
rows,cols = H.shape
background = cv2.resize(background,(cols,rows))

for i in range(rows):
    for j in range(cols):
        if 35 < H[i,j] < 90 and S[i,j] > 50 and V[i,j] < 20:
            img_rgb[i,j] = background[i,j]
            
```

```python
plt.imshow(img_rgb)
```
