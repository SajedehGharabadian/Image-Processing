```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('input/Carrot.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
R,G,B = cv2.split(img)
```

```python
plt.imshow(G,cmap="gray")
```

```python
R_new = B
G_new = G
B_new = B
```

```python
img_result = cv2.merge((B_new,G_new,R_new))
img_result = cv2.cvtColor(img_result,cv2.COLOR_BGR2RGB)
cv2.imwrite('output/green_carrot.jpg',img_result)
```
