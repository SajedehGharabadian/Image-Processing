```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img_orginal = cv2.imread('0.jpg')
gray = cv2.cvtColor(img_orginal,cv2.COLOR_BGR2GRAY)

img = cv2.medianBlur(gray,5)
cnt = 0
circles = cv2.HoughCircles(img, cv2.HOUGH_GRADIENT, 1,15,param1=100,param2=30,minRadius=0,maxRadius=0)
circles = np.uint16(np.around(circles))
for i in circles[0,:]:
    
    cv2.circle(img_orginal,(i[0],i[1]),i[2],(0,255,0),4)
    cnt += 1
cv2.putText(img_orginal,'value = '+ str(cnt),(30,60),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,0),2)
```

```python
plt.imshow(img_orginal,cmap="gray")
```

```python
img_orginal = cv2.imread('1.jpg')

img_orginal1 = img_orginal[:,0:255]
img_orginal2 = img_orginal[:,255:509]

cnt1 = 0
cnt2 = 0

gray = cv2.cvtColor(img_orginal1,cv2.COLOR_BGR2GRAY)

img = cv2.medianBlur(gray,5)

circles = cv2.HoughCircles(img, cv2.HOUGH_GRADIENT, 1,15,param1=80,param2=30,minRadius=3,maxRadius=20)

if circles is not None:
   # Get the (x, y, r) as integers
   circles = np.round(circles[0, :]).astype("int")
   # loop over the circles
   for (x, y, r) in circles:
      cv2.circle(img_orginal1, (x, y), r, (0, 255, 0), 2)
      cnt1 += 1
cv2.putText(img_orginal1,'value = '+ str(cnt1),(30,60),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,0),2)

gray = cv2.cvtColor(img_orginal2,cv2.COLOR_BGR2GRAY)

img = cv2.medianBlur(gray,5)

circles = cv2.HoughCircles(img, cv2.HOUGH_GRADIENT, 1,15,param1=80,param2=30,minRadius=3,maxRadius=20)

if circles is not None:
   # Get the (x, y, r) as integers
   circles = np.round(circles[0, :]).astype("int")
   # loop over the circles
   for (x, y, r) in circles:
      cv2.circle(img_orginal2, (x, y), r, (0, 255, 0), 2)
      cnt2 += 1
cv2.putText(img_orginal2,'value = '+ str(cnt2),(30,60),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,0),2)
```
